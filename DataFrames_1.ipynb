## Drop rows with missing values (NaN)

```python
equity.dropna(how = 'any')
```

```python
equity.dropna(subset=["NAME OF COMPANY","SYMBOL"])
```

## Fill in missing values (NaN)

```python
equity.fillna(0) #Replaces every NaN with provided value in entire df
```

```python
equity["NAME OF COMPANY"] = equity["NAME OF COMPANY"].fillna(0)
```

```python
equity
```

```python
nifty = pd.read_csv("nifty200.csv")
nifty
```

```python
nifty["Close"] = nifty["Close"].astype(int) #float can be converted to int only of there is no NaN.
nifty
```

## Rank values with Rank method

```python
cpi = pd.read_csv("CPIgujarat.csv")
```

```python
#Before ranking, make sure to get rid of missing values. 
```

```python
cpi.info()
```

```python
cpi["Combined"] = cpi["Combined"].astype(int)
```

```python
cpi["Rank"] = cpi["Combined"].rank(ascending = False).astype(int)
```
